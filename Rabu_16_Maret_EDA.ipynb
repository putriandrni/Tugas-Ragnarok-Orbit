### Exploratory Analytics

```python
# Importing Data processing libraries
import numpy as np 
import pandas as pd 
pd.plotting.register_matplotlib_converters()

# Plotting Libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Reading Data
# Download the dataset from here - 
#https://raw.githubusercontent.com/bluedataconsulting/AIMasteryProgram/main/Lab_Exercises/Module3/iris_data.csv
iris=pd.read_csv("https://raw.githubusercontent.com/alien087/pembelajaran-python-AI/main/bank_churn.csv")
```

### Univariative Analysis 

```python
# Distinct Species values
iris["Geography"].unique()
```

```python
# Dividing data
france1 = iris.loc[iris['Geography'] == "France"]
spain1 = iris.loc[iris['Geography'] == "Spain"]
germany1 = iris.loc[iris['Geography'] == "Germany"]
```

```python
france1.head()
```

```python
iris.columns
```

```python
#Ploting the boxplot
sns.boxplot(x = iris['RowNumber'], color='powderblue')
```

```python
# Ploting the KDE
sns.histplot(data=iris['RowNumber'],  color="red", kde=True, stat="density", linewidth=0)  
```

```python
sns.countplot(x=iris['Geography'])
```

### Bivariate Data Analysis

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(5, 5), sharex=True)
sns.despine(left=True)  

# Ploting the histogram with KDE

sns.histplot(data=france1["CreditScore"],label="France",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data=spain1["CreditScore"],label="Spain",kde=True,color='orange',linewidth=2,alpha=0.3)
sns.histplot(data=germany1["CreditScore"],label="Germany",kde=True,color='g',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

* Kelas France rata rata punya besaran CreditScore sebesar 400-800
* Kelas Spain rata rata punya besaran CreditScore sebesar 400-800
* Kelas Germany rata rata punya besaran CreditScore sebesar 400-800

### Multivariate Analysis

```python
sns.pairplot(iris, hue="Geography", diag_kind="hist")
plt.show()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

# Plot the Swarmplot
sns.swarmplot(x=iris['Geography'], y=iris['CreditScore'], ax=axes[0])

# Plot the Boxplot
sns.boxplot(x=iris['Geography'], y=iris['CreditScore'], ax=axes[1])

plt.tight_layout()
```
